```python
# input_dir_dvv is the file location for dvv data folder
# input_dir_dv is the file location for dv data folder

import os
import pandas as pd
import string


input_dir_dvv='Chronus test/Cell C/DVV'
input_dir_dv='Chronus test/Cell C/DV'

dvv_files = [f for f in os.listdir(input_dir_dvv)]# if os.path.isfile(join(input_dir_dvv, f))]
dv_files = [f for f in os.listdir(input_dir_dv)]# if os.path.isfile(join(input_dir_dv, f))]
combined_df=pd.DataFrame()

for dvv_f, dv_f in zip(dvv_files, dv_files):
    if os.stat(input_dir_dvv +'/' +dvv_f).st_size == 0:
        print('File is empty')
    else:
        dvv_df = pd.read_csv(input_dir_dvv +'/' +dvv_f)
        dvv_df = dvv_df[["real_time_106999","real_data_106999","imag_data_106999"]]
        print('dvv len',len(dvv_df))
        if os.stat(input_dir_dv +'/' +dvv_f).st_size == 0:
            print('File is empty')
        else:
            dv_df = pd.read_csv(input_dir_dv + '/'+ dv_f)
            dv_df = dv_df[["real_data_27999999","imag_data_27999999"]]
            # Scaling the dv data to match the dvv data
            dv_df = dv_df.apply(lambda x: x * 10)
            print('dv len:',len(dv_df))

            # Use the combined df data to proceed with analysis. 
            combined_df_cvv_and_cv = pd.concat([dvv_df,dv_df], axis=1)
            combined_df=pd.concat([combined_df,combined_df_cvv_and_cv],axis=0)
combined_df=combined_df.sort_values(by='real_time_106999',ascending=True)
```
